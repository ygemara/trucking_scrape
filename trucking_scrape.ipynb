```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time
from bs4 import BeautifulSoup
import requests

def grab_all_comments(url):
    # Fetch HTML content
    comment_list = []
    html_content = requests.get(url).text

    # Parse HTML
    soup = BeautifulSoup(html_content, "html.parser")

    # Find all blockquote elements
    quote_blocks = soup.find_all("blockquote", class_="messageText SelectQuoteContainer ugc baseHtml")

    # Process each blockquote
    if quote_blocks:
        for quote_block in quote_blocks:
            text = quote_block.text.strip()#.replace("\n","").replace("\t","").replace("\'","'")
            comment_list.append(text)
            comment_list.append("\n")  # Add newline after each comment
    else:
        print("No blockquotes found on the webpage.")
    
    return comment_list

def next_page():
    next_link = WebDriverWait(driver, 10).until(
        EC.element_to_be_clickable((By.XPATH, '//a[@class="text" and contains(text(), "Next")]'))
    )

    # Click the link
    next_link.click()

# Replace 'url' with your starting URL
url = "https://www.thetruckersreport.com/truckingindustryforum/threads/the-worst-shippers-and-receivers-truckers-will-not-buy-their-products.14947"
url = "https://www.thetruckersreport.com/truckingindustryforum/threads/the-best-shippers-and-receivers-truckers-will-buy-their-products.14946/"
# Initialize the WebDriver
driver = webdriver.Chrome()  # Or any other WebDriver

# Navigate to the starting page
driver.get(url)

# Open a file to write comments
with open("comments.txt", "w", encoding="utf-8") as file:
    # Main loop to navigate through pages and scrape information
    while True:
        # Scrape information from the current page
        comments = grab_all_comments(driver.current_url)
        
        for comment in comments:
            print(comment)
            if comment.strip() != "":
                
                file.write(comment)  # Write each comment to the file
                
                file.write("\n")
                file.write("-" * 100)
                file.write("\n")
            #Go to the next page if available, otherwise break out of the loop
        try:
            print(driver.current_url)
            next_page()
            time.sleep(3)  # Adding a delay to allow the next page to load properly
        except Exception as e:
            print("No more pages available. Exiting loop.")
            break

# Close the browser after finishing
driver.quit()
```
